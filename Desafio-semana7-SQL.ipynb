```python
#Criação da tabela clientes
%%sql

CREATE TABLE clientes(id INT PRIMARY KEY, nome VARCHAR(100), idade INT, saldo FLOAT)
```

```python
%%sql
select * from clientes
```

```python
#Selecione o nome e a idade dos clientes com idade superior a 30 anos.
%%sql
select nome, idade from clientes where idade > 30
```

```python
#Calcule o saldo médio dos clientes
%%sql
select avg(saldo) as media from clientes
```

```python
#Encontre o cliente com o saldo máximo.
%%sql


select max(saldo) as maximo from clientes
```

```python
%%sql
select nome from clientes where saldo = 5000.0
```

```python
#Conte quantos clientes têm saldo acima de 1000.
%%sql
select count(id) from clientes where saldo > 1000
```

```python
#Atualize o saldo de um cliente específico.
%%sql
update clientes set saldo = 6000 where id = 3
```

```python
#Remova um cliente pelo seu ID.
%%sql
delete from clientes where id = 6
```

```python
#Criando a tabela compras
%%sql
CREATE TABLE compras (id INT PRIMARY KEY, cliente_id INT, produto VARCHAR(100), valor REAL, FOREIGN KEY(cliente_id) REFERENCES clientes(id));
```

```python
#Inserindo dados na tabela
%%sql
insert into compras (id, cliente_id, produto, valor) values (1, 2, "livro", 56.80), (2, 3, "camisa", 60.0), (3, 4, "panela", 75.0), (4, 6, "toalha", 40.0)
```

```python
%%sql
select * from compras
```

Vamos usar a cláusula JOIN para retornar informações de duas ou mais tabelas combinadas

```python
#Escreva uma consulta para exibir o nome do cliente, o produto e o valor de cada compra.
%%sql
SELECT nome, produto, valor FROM compras INNER JOIN clientes ON clientes.id = compras.cliente_id
```
